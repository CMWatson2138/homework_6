```python
#B
#In order to clean the data we decided to check for duplicates and change the capitalization of the Categories column. We also added columns for the month and year of air date.
```

```python
clean_jeopardy=jeopardy_df.drop_duplicates()
clean_jeopardy[" Category"]=clean_jeopardy[" Category"].str.capitalize()
```

```python
clean_jeopardy['Year'] = pd.DatetimeIndex(clean_jeopardy[' Air Date']).year
clean_jeopardy['Month'] = pd.DatetimeIndex(clean_jeopardy[' Air Date']).month
clean_jeopardy.head()
```

```python
#c. How many unique shows were there each year?
#group by year aired
shows_per_year=clean_jeopardy.groupby(["Year"])
#count unique shows by number
unique_shows=shows_per_year.aggregate({"Show Number": "nunique"})
#rename column
unique_shows.rename(columns={"Show Number": "Unique Shows per Year"}, inplace=True)
display(unique_shows)
```

```python
#d. How many times was each category used?
#First group data by category
cat_df=clean_jeopardy.groupby(" Category")
#Next count number of times a category appears
cat_pop=cat_df.aggregate({" Category":"count"})
cat_pop
```

```python
#e. Which month (over the years) has had the most shows?

#First group data by month
month_df=clean_jeopardy.groupby("Month")
#Next count number of shows in each month
month_count_df=month_df.aggregate({"Show Number": "nunique"}) 
#rename column
month_count_df.rename(columns={"Show Number": "Number of Shows"}, inplace=True)
month_count_df
```

```python
# f. Are there any answers that appeared multiple times? Find the questions for the top three most common answers.

#First group by the answers column
most_pop_answers=clean_jeopardy.groupby(" Answer")
#Next count the number of times as answer appears
most_pop_ans=most_pop_answers.aggregate({" Answer": "count"})
#Rename column
most_pop_ans.rename(columns={" Answer": "Number of Times Answer Appeared"}, inplace=True)
#Sort by number of answers in descending order
most_popular_sorted=most_pop_ans.sort_values("Number of Times Answer Appeared", ascending=False)
#Print top three answers
print(most_popular_sorted.head(3))

#I really couldn't figure this one out
#Retrieve questions associated with top three answers
#for index, row in jeopardy_df.iterrows():
    #if jeopardy_df.loc[(jeopardy_df[' Answer'] == 'China') | (jeopardy_df[' Answer'] == 'Australia')| (jeopardy_df[' Answer'] == 'Japan')]:
        #print(f'Index: {index}, Question: {row.get(" Question", 0)}')
    #else:
        #pass 
```

```python
#2. Create a visualization that shows the top 5 categories (those referenced the most often). Be sure to label appropriately

#What are the top five categories on the show?
#Answer: Before & After, Science, Literature, American History, and Potpourri

#First group data by category
most_pop_df=clean_jeopardy.groupby(" Category")
#Next count number of times a category appears
most_popular=most_pop_df.aggregate({" Category":"count"})
#rename column
most_popular.rename(columns={" Category": "Number_of_Questions"}, inplace=True)
#sort by descending order
most_popular_sorted=most_popular.sort_values("Number_of_Questions", ascending=False)
#slice first five rows
most_pop_vis=most_popular_sorted.iloc[0:5, :]
most_pop_vis
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator, FixedFormatter


fig, ax = plt.subplots()
ax.bar(most_pop_vis.index, most_pop_vis["Number_of_Questions"])
ax.set_xticks(most_pop_vis.index)
ax.set_xticklabels(most_pop_vis.index, rotation=90)
ax.set_ylabel("Number_of_Questions")
plt.show()
```
